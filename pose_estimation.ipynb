## defining calculate angle function to calculate the angles

```python
def calculate_angle(a,b,c):
    a = np.array(a) # First
    b = np.array(b) # Mid
    c = np.array(c) # End

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle >180.0:
        angle = 360-angle

    return angle
```

```python
import cv2
import numpy as np
import mediapipe as mp
from google.colab.patches import cv2_imshow

# Define the angle calculation function
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360 - angle

    return angle

# Load image
img = cv2.imread("/content/drive/MyDrive/HW1/Images/Mountain.jpg")

# Check if image loaded
if img is None:
    print("Image not found!")
else:
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img_rgb.flags.writeable = False

        results = pose.process(img_rgb)

        img_rgb.flags.writeable = True
        img3 = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)

        try:
            landmarks = results.pose_landmarks.landmark

            right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                          landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
            right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                           landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
            right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                         landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
            right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                              landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                           landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]

            angle = round(calculate_angle(right_ankle, right_knee, right_hip), 2)
            angle_one = round(calculate_angle(right_hip, right_shoulder, right_elbow), 2)

            print("Knee angle:", angle, "| Shoulder angle:", angle_one)

            mp_drawing.draw_landmarks(img3, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                      mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                                      mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))

            right_knee_coord = tuple(np.multiply(right_knee, [img.shape[1], img.shape[0]]).astype(int))
            right_shoulder_coord = tuple(np.multiply(right_shoulder, [img.shape[1], img.shape[0]]).astype(int))

            cv2.putText(img3, str(angle), right_knee_coord,
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 2, cv2.LINE_AA)

            cv2.putText(img3, str(angle_one), right_shoulder_coord,
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 2, cv2.LINE_AA)

        except Exception as e:
            print("Landmark detection failed:", e)

        # Display the final annotated image
        cv2_imshow(img3)
```

## GODDESS

```python
import cv2
import numpy as np
import mediapipe as mp
from google.colab.patches import cv2_imshow

# Define the angle calculation function
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians * 180.0 / np.pi)

    if angle > 180.0:
        angle = 360 - angle

    return angle

# Load the image
img = cv2.imread("/content/drive/MyDrive/HW1/Images/Goddess.jpg")

# Validate the image
if img is None:
    print(" Image not found. Please check the path.")
else:
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img_rgb.flags.writeable = False

        results = pose.process(img_rgb)

        img_rgb.flags.writeable = True
        img4 = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)

        try:
            landmarks = results.pose_landmarks.landmark

            # Right side
            right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                          landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
            right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                           landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
            right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                         landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
            right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                              landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
            right_elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                           landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]

            # Left side
            left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                          landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
            left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                         landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
            left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
                        landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
            left_elbow = [landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].x,
                          landmarks[mp_pose.PoseLandmark.LEFT_ELBOW.value].y]
            left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                             landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]

            # Calculate angles
            angle = round(calculate_angle(right_ankle, right_knee, right_hip), 2)
            angle_one = round(calculate_angle(right_hip, right_shoulder, right_elbow), 2)
            angle_two = round(calculate_angle(left_ankle, left_knee, left_hip), 2)
            angle_three = round(calculate_angle(left_hip, left_shoulder, left_elbow), 2)

            print("Angles (degrees):")
            print("Right Leg:", angle)
            print("Right Arm:", angle_one)
            print("Left Leg:", angle_two)
            print("Left Arm:", angle_three)

            # Draw pose
            mp_drawing.draw_landmarks(img4, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                      mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2),
                                      mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))

            # Annotate image with angles
            h, w, _ = img.shape  # original image dimensions
            points = {
                tuple(np.multiply(right_knee, [w, h]).astype(int)): angle,
                tuple(np.multiply(right_shoulder, [w, h]).astype(int)): angle_one,
                tuple(np.multiply(left_knee, [w, h]).astype(int)): angle_two,
                tuple(np.multiply(left_shoulder, [w, h]).astype(int)): angle_three,
            }

            for coord, ang in points.items():
                cv2.putText(img4, str(ang), coord,
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)

        except Exception as e:
            print(" Pose landmarks not detected:", e)

        # Display final image
        cv2_imshow(img4)
```
